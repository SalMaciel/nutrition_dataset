#### At first glance
All data is surrounded by **~** with the exception of NaN values and some numeric columns. **~** is part of the access type files but not necessary for this project.

```python
test = pd.DataFrame()
for rows in data_src_df:
    test[rows] = data_src_df[rows].str.replace("~","")
test.head()
```

```python
data_src_df.dtypes
```

```python
#Trying a new way to remove ~ from data sets
test1 = WEIGHT_df[0].str
test1 = np.where(test1.startswith('~'), test1.replace('~', '', regex = True),\
                 np.where(test1.endswith('~'), test1.replace('~', '', regex = True), test1))
test1
```

```python
test3 = WEIGHT_df[3].str
test3 = np.where(test3.startswith('~'), test3.replace('~', '', regex = True),\
                 np.where(test3.endswith('~'), test3.replace('~', '', regex = True), test3))
test3
```

Both of these methods don't work on numeric type columns

```python
try_new_df = pd.read_csv("data/WEIGHT.txt", sep = "^", header = None, dtype = "object",encoding = "ISO-8859-1")
try_new_df.head()
```

```python
try_new_df.dtypes
```

```python
try_new_test = pd.DataFrame()
for rows in try_new_df:
    try_new_test[rows] = try_new_df[rows].str.replace("~","")
try_new_test.head()
```

Reading all columns as type object makes it easy to remove '~' from data
